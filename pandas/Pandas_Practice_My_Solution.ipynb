## Part 2 – Statistics

```python
avg_price = df_data['price'].mean()
max_price = df_data['price'].max()
avg_price, max_price
```

```python
avg_price, avg_num_items = df_data[['price','num_item']].mean()
avg_price, avg_num_items
```

```python
df_data['total'] = df_data['price'] * df_data['num_item']
df_data.head()
```

```python
df_data.sort_values('total', ascending=False).head(3)['total'].mean()
```

```python
df_data['product_id'].nunique()
```

```python
desc = df_data['total'].describe()
desc['mean'] - desc['50%']
```

```python
df_data.drop(columns=['total'], inplace=True)
```

```python
df_data.info()
```

```python
df_data['invoice_time'] = pd.to_datetime(df_data['invoice_time'])
df_data['DOW'] = df_data['invoice_time'].dt.day_name()
df_data['Month'] = df_data['invoice_time'].dt.month
df_data.head()
```

## Part 3 – Titanic Data

```python
df_titanic = pd.read_csv('Data/Titanic.csv')
df_titanic.head()
```

```python
df_titanic.rename(columns={'Name':'Passenger_Name','Ticket':'Ticket_Num'}, inplace=True)
```

```python
df_titanic.set_index('PassengerId', inplace=True)
df_titanic.loc[896, 'Passenger_Name']
```

```python
df_titanic.reset_index(inplace=True)
```

```python
df_titanic['Age'].isnull().sum()
```

```python
age_avg = df_titanic['Age'].mean()
df_titanic['Age'].fillna(age_avg, inplace=True)
```
